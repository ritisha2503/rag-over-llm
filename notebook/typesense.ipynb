```python
loader = TextLoader("../test.txt")
documents = loader.load()
text_splitter = CharacterTextSplitter(chunk_size = 1000, chunk_overlap = 100)
docs = text_splitter.split_documents(documents)

embeddings = HuggingFaceEmbeddings()
```
